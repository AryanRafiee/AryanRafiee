## Data Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
X = df.iloc[:,:].values
```

```python
X
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X = imputer.fit_transform(X)
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
X1 =X.T
X1
```

## converting Data format to TSLearn Time Series

```python
from tslearn.utils import to_time_series_dataset
```

```python
X2 = to_time_series_dataset(X1)
```

## Clustering Using Euclidean Metric

```python
from tslearn.clustering import TimeSeriesKMeans
km =TimeSeriesKMeans(n_clusters=3, metric="euclidean", random_state=0)
km.fit_predict(X2)
```

## Clustering Using softdtw Metric

```python
km2 =TimeSeriesKMeans(n_clusters=3, metric="softdtw", random_state=0)
km2.fit_predict(X2)
```
