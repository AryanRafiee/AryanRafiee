# Peak Forcasting Using Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = pd.read_csv('1402-1.csv')
X = df.iloc[:,[2,3]].values
y = df.iloc[:,-1].values
df.head()
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.hist(y)
plt.show()
```

```python
plt.scatter(df['Max Temp'], df['Peak'])

plt.xlabel("Temperature") #x label
plt.ylabel("Peak")
plt.show()
```

```python
plt.scatter(df['Max Temp'], df['Peak'])
plt.xlabel("Temperature") #x label
plt.ylabel("Peak")
plt.show()
```

```python
plt.scatter(df['U Max'], df['Peak'])
plt.xlabel("humidity") #x label
plt.ylabel("Peak")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X_train)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y_train)
```

```python
y_pred = lin_reg.predict(poly_reg.transform(X_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
print (np.concatenate((y_pred.reshape(len (y_pred),1), y_test.reshape(len (y_test),1)),1))
```

```python
Today_Peak_prediction=lin_reg.predict(poly_reg.transform([[47,14]]))
print(Today_Peak_prediction)
```

```python
plt.scatter(t,y_test, c = 'blue')
plt.scatter(t,y_pred, c ='red')
plt.title("Original Peak VS Predicted Peak") #title
plt.xlabel("Temperature") #x label
plt.ylabel("Peak") #y label

plt.show()
```
