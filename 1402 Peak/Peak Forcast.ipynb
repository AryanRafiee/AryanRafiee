```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1402-1.csv")
df.head()
df.shape
```

```python
df.head()
```

```python
X = df.iloc[:,[2,3]].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(df['Min Temp'], df['Peak'])
plt.show()
```

```python
plt.scatter(df['U Max'], df['Peak'])
plt.show()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=100 , activation= 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=100 , activation= 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=90 , activation= 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=80, activation= 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=60 , activation= 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(optimizer='adam' , loss = 'MAE')
```

```python
ann.fit(X_train,y_train, batch_size=32 , epochs = 500) 
```

```python
y_pred = ann.predict(X_test)
```

```python
print (np.concatenate((y_pred.reshape(len (y_pred),1), y_test.reshape(len (y_test),1)),1))
```

```python
from sklearn.metrics import  r2_score
r2_score(y_test, y_pred)
```

```python
y_hat = ann.predict(X_train)
r2_score(y_train,y_hat)
```
